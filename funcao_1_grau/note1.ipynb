# Função do 1° Grau

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Definição

Chame-se função do 1° grau a função $f: \mathbb{R} \rightarrow \mathbb{R}$ definida por $f(x) = ax + b$, com a e b números reais. a é o coeficiente angular da reta e determina sua inclinação, b é o coefieciente linear da reta e determina a intersecção da reta com o eixo y.

#### Representação gráfica de $y = 100 + 0,05x$
Uma loja de eletrodomésticos contrata vendedores com as seguin-
tes condições salariais: uma quantia fixa de R$ 100,00 mais 5% sobre
as vendas efetuadas.

```python
# Cria um array de valores da area do grafico
x = np.linspace(0, 5000, 100)

# Define a função y = 100 + 0,05
y = 100 + 0.05 * x


# Cria o gráfico
plt.plot(x, y)

# Define labels para os títulos dos gráficos
plt.title('y = 100 + 0,05x')
plt.xlabel('x')
plt.ylabel('y')

# Ativa a grade no gráfico
plt.grid(True)

plt.show()
```
